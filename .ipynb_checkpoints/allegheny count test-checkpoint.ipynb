```python
url = 'http://www.alleghenycounty.us/emergency-services/police-departments.aspx'
r = http.request('GET', url)
print(r.status)
if r.status == 200:
    soup = BeautifulSoup(r.data, 'lxml')
```

```python
dept_dict = {}
for p in soup.find_all('tr'):
    l = []
    for s in p.find_all('td'):
        if s.find('b') != None:
            entry = re.sub('\s+', ' ', s.find('b').text).strip()
        elif s.find('strong') != None:
            entry = re.sub('\s+', ' ', s.find('strong').text).strip()
        else:
            entry = re.sub('\s+', ' ', s.text).strip()
        l.append(entry.replace('P. D.', 'P.D.'))
    if len(l) != 0:
        agency = l[0]
        jurisdiction = l[1].replace('P.D.', '').strip()
        if 'Covered by' in jurisdiction:
            continue
        else:
            if len([w for w in jurisdiction.split() if w in ['Borough', 'Township', 'Village', 'Twp.']]) > 1:
                listo = []
                temp = []
                for w in jurisdiction.split():
                    if w not in ['Borough', 'Township', 'Village']:
                        temp.append(w)
                    else:
                        temp.append(w)
                        listo.append(' '.join(temp))
                        temp = []
                dept_dict[l[0]] = listo
            else:
                dept_dict[l[0]] = l[1]
```

```python
northern = ['Pine Township', 'Bradford Woods Borough', 'Marshall Township', 'Richland Township']
for d in northern:
    dept_dict.pop(d, None)
dept_dict['Northern Regional P.D.'] = northern
```

```python
juris_dict = {}
for k, v in dept_dict.items():
    if type(v) == str:
        juris_dict[v] = k
    else:
        for i in v:
            juris_dict[i] = k
```

```python
df = pd.DataFrame(juris_dict, index=['agency']).T.reset_index().rename(columns={'index':'jurisdiction'})
```

```python
df['jurisdiction'] = df['jurisdiction'].map(lambda x: x.replace('P.D.', '').strip())
```

```python
df
```

```python
def juris_cleaner(j):
    j2 = j.lower()
    replacers = {'city of':'', '(westmoreland county)':'', 'mt.':'mount',
                 'greentree':'green tree', 'ohara':'o\'hara', 'alleppo':'aleppo'}
    for k, v in replacers.items():
        j2 = j2.replace(k, v).strip()
    j3 = [w for w in j2.split() if w not in ['borough', 'township', 'municipality', 'city']]
    return ' '.join(j3)
```

```python
df['name'] = df['jurisdiction'].map(juris_cleaner)
```

```python
ag_name = [x.lower() for x in ag['NAME']]
juris_name = [x for x in df['name']]
```

```python
ag['NAME'] = ag['NAME'].map(lambda x: x.lower())
```

```python
ag_pd = pd.merge(ag, df[['name', 'agency']], how='outer', left_on='NAME', right_on='name').drop('name', axis=1).fillna('None')
```

```python
j_size = {}
for k,v in dept_dict.items():
    if type(v) == list:
        j_size[k] = str(len(v))
    else:
        j_size[k] = str(1)
```

```python
ag_pd['juris size'] = ag_pd['agency'].map(lambda x: j_size[x] if x != 'None' else 'no data')
```

```python
ag_pd.plot(figsize=(15,15), edgecolor='black', column='juris size', alpha=0.7,
           legend=True, legend_kwds={'loc': 4})
plt.legend(loc='lower right')
plt.title('Jurisdiction Coverage by Allegheny County Police Departments', size=16)
plt.axis('off');
```

```python
def pop_over_juris(row):
    try:
        return row['pop 2017']/int(row['juris size'])
    except ValueError:
        return np.NaN
```

```python
ag_pd_pop['ppl per PD'] = ag_pd_pop.apply(pop_over_juris, axis=1)
```

```python
ag_pd_pop.dropna().plot(figsize=(15,15), edgecolor='black', column='ppl per PD', alpha=0.7,
              legend=True, legend_kwds={'loc': 4})
plt.axis('off');
```

```python
fig = plt.figure(figsize=(12,10))

plt.subplots_adjust(left=None, bottom=None, right=None, top=1.2,
                wspace=None, hspace=.3)

p = 1
for n in [5000, 10000, 50000, 400000][::-1]:
    plt.subplot(2,2,p)
    plt.title(s=f'people per PD, population < {n}')
    data = ag_pd_pop[ag_pd_pop['pop 2017'] < n]
#     plt.scatter(data=data, x='pop 2017', y='ppl per PD')
    sns.regplot(data=data, x='pop 2017', y='ppl per PD')
    plt.xlim(-10,n)
    plt.xticks(rotation=45)
    p += 1;
```
